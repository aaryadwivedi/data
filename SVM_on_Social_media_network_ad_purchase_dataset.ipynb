```
# Creating Input and Output Variables
X = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```
# It time to split the data into testing and training 
```

```
# 75% data in training and 25% for testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```
# Creating the model with help of SVM's SVC 
from sklearn.svm import SVC
model  = SVC()
model.fit(X_train, y_train)
```

```
# Testing the score of model
model.score(X_test,y_test)
```

```
# Predicting the result
```

```
# model.predict([[Age, EstSalary]])
model.predict([[29, 120000]])
```

```
model2  = SVC(kernel='linear')
model2.fit(X_train, y_train)
```

```
# Testing the score of model2
model2.score(X_test,y_test)
```

```
model3  = SVC(kernel='sigmoid') # poly, sigmoid, 
model3.fit(X_train, y_train)
# Testing the score of model2
model3.score(X_test,y_test)
```

```
model3  = SVC(kernel='linear',probability=True) # poly, sigmoid, 
model3.fit(X_train, y_train)
# Testing the score of model2
model3.score(X_test,y_test)
```
